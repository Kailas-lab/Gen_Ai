**For CSV file loaders**

```python
from langchain.document_loaders import CSVLoader

loader = CSVLoader(file_path="zero.csv")
docs = loader.load()
print(docs)
```

**For web extraction**

```python
from langchain.document_loaders import WebBaseLoader

headers = {"User-Agent": "Mozilla/5.0"}
loader = WebBaseLoader("https://en.wikipedia.org/wiki/List_of_asset_management_firms")
docs = loader.load()
print(docs[0].page_content)  # Print page content
```
